```python
print("=== Employee Management System ===")

class Employee:
    def __init__(self, eid, name, age, salary):
        self.eid = eid
        self.name = name
        self.age = age
        self.salary = salary

    def show(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"ID: {self.eid}")
        print(f"Salary: {self.salary}")


class Manager(Employee):
    def __init__(self, eid, name, age, salary, dept):
        super().__init__(eid, name, age, salary)
        self.dept = dept

    def show(self):
        super().show()
        print(f"Department: {self.dept}")


class Developer(Employee):
    def __init__(self, eid, name, age, salary, lang):
        super().__init__(eid, name, age, salary)
        self.lang = lang

    def show(self):
        super().show()
        print(f"Language: {self.lang}")


def menu():
    records = []

    while True:
        print("1. Add Employee")
        print("2. Add Manager")
        print("3. Add Developer")
        print("4. Show All")
        print("5. Exit")

        ch = input("Enter choice: ")

        if ch in ["1", "2", "3"]:
            eid = input("Enter ID: ")
            name = input("Enter Name: ")
            age = int(input("Enter Age: "))
            salary = int(input("Enter Salary: "))

            if ch == "1":
                records.append(Employee(eid, name, age, salary))

            elif ch == "2":
                dept = input("Enter Department: ")
                records.append(Manager(eid, name, age, salary, dept))

            elif ch == "3":
                lang = input("Enter Language: ")
                records.append(Developer(eid, name, age, salary, lang))

        elif ch == "4":
            if not records:
                print("No records found")
            else:
                for obj in records:
                    print("-" * 20)
                    obj.show()

        elif ch == "5":
            print("Exiting program...")
            break

        else:
            print("Invalid choice")


if __name__ == "__main__":
    menu()
```
